```python
import tensorflow as tf
import numpy as np
import pandas as pd
from utils.network import *
from sklearn.metrics import confusion_matrix, precision_score, recall_score, precision_recall_curve
```

```python
X_train=np.loadtxt('data/X_train.csv',delimiter=',')
X_test=np.loadtxt('data/X_test.csv',delimiter=',')
y_train=np.loadtxt('data/y_train.csv',delimiter=',')
y_test=np.loadtxt('data/y_test.csv',delimiter=',')
```

```python
model=mlp(True,hidden_layer=[40,10])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
y_train=y_train.astype('int32')
y_test=y_test.astype('int32')
```

```python
y_train = np.eye(10)[y_train.reshape(-1)]
y_test = np.eye(10)[y_test.reshape(-1)]
```

```python
history = model.fit(x=X_train,
                    y=y_train,
                    batch_size=400,
                    epochs=500,
#                     initial_epoch=8,
#                    callbacks=[cp_callback],
#                    validation_data=(X_val, y_val),
                    verbose=1)
```

```python
import os

if not os.path.exists('/training'):
    os.mkdir('./training')
checkpoint_path = "training/cp-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

# create callback to store weights after each epoch
cp_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_path, 
    verbose=1, 
    save_weights_only=True,
    period=1)
```

```python
model.save_weights(checkpoint_path.format(epoch=0))
```

```python
model.load_weights('training/cp-0000.ckpt')
```

```python
y_pred=np.argmax(model.predict(X_test),axis=1)
```

```python
y_test=np.loadtxt('data/y_test.csv',delimiter=',')
```

```python
confusion_matrix(y_test,y_pred)
```

```python
precision_score(y_test,y_pred,average='weighted')
```

```python
recall_score(y_test,y_pred,average='weighted')
```

```python
np.average(y_pred==y_test)
```
